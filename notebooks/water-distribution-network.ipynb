# Exeter water distribution network

Visual representation of the water distribution network of the city of Exeter, UK.

```python
import folium as fl
import networkx as nx
import matplotlib.pyplot as plt
from pathlib import Path
from net_flow_opt.utils import read_inp
```

```python
data = read_inp(Path('..', 'data', 'systems_data', 'exeter.inp'))
```

```python
G = nx.Graph()
fig, ax = plt.subplots(1, 1, figsize=(20, 20))
# add nodes
for i, x, y in data['coordinates'][['X-Coord', 'Y-Coord']].to_records():
    G.add_node(i, pos=(x, y) )
# add edges
for i, j, ID, l, d, r, ml, s in data['pipes'].to_records():
    G.add_edge(i, j)

fig = nx.draw(G, nx.get_node_attributes(G, 'pos'), with_labels=False, node_size=2)

plt.savefig('exeter-wdn', bbox_inches='tight')
```
